```python
X=df.drop('loan_payment',axis=1).values
y=df['loan_payment'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.callbacks import EarlyStopping
```

```python
early_stopping=EarlyStopping(monitor='val_loss',patience=20,verbose=3)
```

```python
model=Sequential()
```

```python
model.add(Dense(53,activation='relu'))

model.add(Dense(26,activation='relu'))
model.add(Dense(13,activation='relu'))
model.add(Dense(6,activation='relu'))

model.add(Dense(1,activation='sigmoid'))

model.compile(optimizer='adam',loss='binary_crossentropy')
```

```python
model.fit(X_train,y_train,epochs=600,validation_data=(X_test,y_test),callbacks=[early_stopping],verbose=3)
```

```python
losses=model.history.history
```

```python
loss=pd.DataFrame(losses)
loss
```

```python
loss.plot()
```

```python
y_hat=model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,y_hat))
```

```python
print(classification_report(y_test,y_hat))
```
