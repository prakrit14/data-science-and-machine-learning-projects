#### importing all the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
```

#### loading the dataset

```python
data=pd.read_csv('data.csv',error_bad_lines=False)
data.head()
```

#### dropping missing values from the dataset

```python
data.dropna(inplace=True)
```

#### plot for distribution of strength of password

```python
plt.style.use(['ggplot'])
sns.countplot(data['strength'])
```

#### extracting the password column and saving them in a tuple

```python
password_tuple=np.array(data)
```

```python
password_tuple
```

#### splitting data into x and y (dependant and independant variables)

```python
x=[labels[0] for labels in password_tuple]
y=[labels[1] for labels in password_tuple]
```

```python
x
```

```python
y
```

#### creating a custom function to split input into characters of list

```python
def word_divide_char(inputs):
    character=[]
    for i in inputs:
        character.append(i)
    return character
```

```python
word_divide_char('kzde5577')
```

#### import TF-IDF vectorizer to convert String data into numerical data

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer=TfidfVectorizer(tokenizer=word_divide_char)
```

#### apply TF-IDF vectorizer on data

```python
X=vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names()
```

#### checking if the vectorizer works 

```python
first_document_vector=X[0]
first_document_vector
```

```python
first_document_vector.T.todense()
```

```python
df=pd.DataFrame(first_document_vector.T.todense(),index=vectorizer.get_feature_names(),columns=['TF-IDF'])
df.sort_values(by=['TF-IDF'],ascending=False)
```

#### splitting data into train & test
    train---> To learn the relationship within data, 
    test-->  To do predictions, and this testing data will be unseen to my model

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

#### using logistic regression

```python
clf=LogisticRegression(random_state=0,multi_class='multinomial')
```

```python
clf.fit(X_train,y_train)
```

#### doing prediction for specific custom data

```python
dt1=np.array(['%@123abcd'])
pred=vectorizer.transform(dt1)
clf.predict(pred)
```

```python
dt2=np.array(['%@123abcd8954utgrg4f4&^&%$#@$@$hbd34f'])
pred=vectorizer.transform(dt2)
clf.predict(pred)
```

```python
dt3=np.array(['abcde'])
pred=vectorizer.transform(dt3)
clf.predict(pred)
```

####  predicting on X_Test data

```python
y_pred=clf.predict(X_test)
y_pred
```

#### checking metrics of our model using confusion_matrix,accuracy_score and classification report

```python
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(accuracy_score(y_test,y_pred))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

#### using random forest classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
y_hat=rfc.predict(X_test)
```

#### checking metrics of our model using confusion_matrix,accuracy_score and classification report

```python
print(confusion_matrix(y_hat,y_pred))
```

```python
print(classification_report(y_hat,y_pred))
```

```python
print(accuracy_score(y_hat,y_pred))
```

#### evidently both our models do fairly well to classify the strength of the password
